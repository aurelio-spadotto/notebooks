```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
R = 3.91e-6
C0 = 0.81e-6
C2 = 7.83e-6
C4 = -4.39e-6
P = 0
A = 0
fig,ax = plt.subplots (figsize=(10,10))
ax.axis('equal')

for ino in range(N):
    theta1 = 2*np.pi/N*ino
    theta2 = 2*np.pi/N*(ino+1)
    x1 = R*np.cos(theta1)
    y1 = (-1+ 2*(np.sin(theta1)>0))*np.sqrt(1-(x1/R)**2)*(C0 + C2*(x1/R)**2 + C4*(x1/R)**4)
    x2 = R*np.cos(theta2)
    y2 = (-1+ 2*(np.sin(theta2)>0))*np.sqrt(1-(x2/R)**2)*(C0 + C2*(x2/R)**2 + C4*(x2/R)**4)
    ds = np.sqrt((x2-x1)**2+(y1-y2)**2)
    P = P + ds
    da = 0.5*np.sqrt(x1**2+y1**2)*np.sqrt(x2**2+y2**2)*np.sin(2*np.pi/N)
    A = A + da
    ax.scatter(x1,y1)

print ('Perimeter', P)
print ('Area', A)
```

```python
red_area = A/((P/2/np.pi)**2*np.pi)
print('Reduced Area: ',red_area)
print ('R_eff', P/2/np.pi)
```
