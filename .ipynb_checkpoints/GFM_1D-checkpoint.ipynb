```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from scipy.sparse.linalg import bicgstab 
```

```python
def error_vs_h(N):
    ###################################
    #Calculate errors of GFM          # 
    ###################################
    #data 
    x = np.linspace(-0.5,0.5,N+1) #node coordinates 
    u = np.zeros(N+1)        #solution vector
    f = np.zeros(N+1)        #rhs vector
    M = np.zeros((N+1,N+1))    #system matrix
    a = -0.2475              #jump position
    b =  0.2475
    V = 1.0                #jump
    CR = 0.1               #conductivity ratio
    h = 1/N                #step
    #determine nodes to correct
    for i in range(1,N):
        if (x[i]<a and x[i+1]>a):
            i_plus_a = i
        if (x[i]>a and x[i-1]<a):
            i_minus_a = i
    for i in range(1,N-1):
        if (x[i]<b and x[i+1]>b):
            i_minus_b = i
        if (x[i]>b and x[i-1]<b):
            i_plus_b = i
            
    #calculate chi
    chi_minus_a  = -(x[i_plus_a]-a)*(1/CR-1)/h
    chi_plus_a   = -(x[i_minus_a]-a)*(1-CR)/h
    chi_minus_b  = (x[i_plus_b]-b)*(1/CR-1)/h
    chi_plus_b   = (x[i_minus_b ]-b)*(1-CR)/h
    #print ("diffs", x[i_plus_a]-a, x[i_minus_a]-a)
    
    #assemble laplacian operator
    for i in range(0,N+1):
        for j in range(0,N+1):
            if (j-i==1):
                M[i,j] = 1/h**2
            elif (j-i==0):
                M[i,j] = -2/h**2
            elif (j-i==-1):
                M[i,j] = 1/h**2
    #force Dirichlet conditions
    M[0,:]=np.zeros(N+1)
    M[N,:] = np.zeros(N+1)
    M[0,0] = 1.0
    M[N,N]=1.0
    f[0] = 0.0
    f[N] = 1.0
    #GFM correction
    M[i_minus_a,i_minus_a] = M[i_minus_a,i_minus_a]+(1/h**2)*(chi_minus_a)/(1+chi_minus_a)
    M[i_minus_a,i_plus_a] = M[i_minus_a, i_plus_a] +(1/h**2)*(-chi_minus_a)/(1+chi_minus_a)
    f[i_minus_a] = f[i_minus_a] +(1/h**2)*(1/(1+chi_minus_a))*V
    M[i_plus_a,i_plus_a] = M[i_plus_a,i_plus_a] +(1/h**2)*(chi_plus_a)/(1+chi_plus_a)
    M[i_plus_a,i_minus_a] = M[i_plus_a, i_minus_a]+(1/h**2)*(-chi_plus_a)/(1+chi_plus_a)
    f[i_plus_a] =  f[i_plus_a] -(1/h**2)*(1/(1+chi_plus_a))*V
    M[i_minus_b,i_minus_b] = M[i_minus_b,i_minus_b]+(1/h**2)*(chi_minus_b)/(1+chi_minus_b)
    M[i_minus_b,i_plus_b] = M[i_minus_b, i_plus_b] +(1/h**2)*(-chi_minus_b)/(1+chi_minus_b)
    f[i_minus_b] = f[i_minus_b] +(1/h**2)*(1/(1+chi_minus_b))*V
    M[i_plus_b,i_plus_b] = M[i_plus_b,i_plus_b] +(1/h**2)*(chi_plus_b)/(1+chi_plus_b)
    M[i_plus_b,i_minus_b] = M[i_plus_b, i_minus_b]+(1/h**2)*(-chi_plus_b)/(1+chi_plus_b)
    f[i_plus_b] =  f[i_plus_b] -(1/h**2)*(1/(1+chi_plus_b))*V 
    """"
    print ("ratios", 1/CR -1, 1-CR)
    print ("chi", chi_minus_b, chi_plus_b)
    print ("rhs", f[i_minus_b], f[i_plus_b])
    print ("cut coeffs", M[i_minus_b, i_plus_b]-1/h**2, M[i_plus_b, i_minus_b]-1/h**2)
    """
    #system inversion
    u = np.linalg.solve(M,f)
    #errors
    err_u_jump = abs(u[i_plus_a]-u[i_minus_a] - V) + abs(u[i_plus_b]-u[i_minus_b] - V)
    err_u_der_jump = abs(CR/0.2525*(u[i_plus_a]-u[0])-1/0.495*(u[i_minus_b]-u[i_minus_a])) + \
                     abs(CR/0.2525*(u[N]-u[i_plus_b])-1/0.495*(u[i_minus_b]-u[i_minus_a]))
   

    
   # edge operators
    S = np.ones(N)/h**2
    T = np.zeros(N)
    D = np.zeros(N+1)
    A_cut_in_a  =+(1/h**2)*(-chi_minus_a)/(1+chi_minus_a)
    A_cut_out_a =+(1/h**2)*(-chi_plus_a)/(1+chi_plus_a)
    A_cut_in_b  =+(1/h**2)*(-chi_minus_b)/(1+chi_minus_b)
    A_cut_out_b =+(1/h**2)*(-chi_plus_b)/(1+chi_plus_b)
    A_diag_minus_a =+(1/h**2)*(chi_minus_a)/(1+chi_minus_a)
    A_diag_plus_a  =+(1/h**2)*(chi_plus_a)/(1+chi_plus_a)
    A_diag_minus_b =+(1/h**2)*(chi_minus_b)/(1+chi_minus_b)
    A_diag_plus_b  =+(1/h**2)*(chi_plus_b)/(1+chi_plus_b)
    #print ("node coeffs", A_cut_in_a, A_cut_out_a, A_cut_in_b, A_cut_out_b)
    #correction
    S[i_plus_a] = S[i_plus_a] + 0.5*(A_cut_in_a + A_cut_out_a)
    T[i_plus_a] = T[i_plus_a] + 0.5*(A_cut_out_a-A_cut_in_a)    
    S[i_minus_b] = S[i_minus_b] + 0.5*(A_cut_in_b+A_cut_out_b)
    T[i_minus_b] = T[i_minus_b] + 0.5*(A_cut_in_b-A_cut_out_b)
    D[i_minus_a]  = A_diag_minus_a + A_cut_out_a  
    D[i_plus_a]   = A_diag_plus_a  + A_cut_in_a 
    D[i_minus_b]  = A_diag_minus_b + A_cut_out_b
    D[i_plus_b]   = A_diag_plus_b  + A_cut_in_b
    #print (S[i_minus_b], T[i_minus_b])
    
    """""
    #CG loop 
    u = np.zeros(N+1)
    u[0] = 0
    u[N] = 1
    r = f - mat_vec_prod(D,S,T,u)
    r[0] = r[N] = 0
    p =r
    rho_0 = np.dot(r,r)
    #theta = np.dot(p, mat_vec_prod(D,S,T,p))
    theta = np.dot(p, M@p)
    it = 0
    maxiter = 50
    while (rho_0 > 1e-12 and it<maxiter):
        alpha = rho_0/theta
        u = u + alpha*p 
        u[0] = 0
        u[N] = 1
        #r = r - alpha*mat_vec_prod(D,S,T,p)
        r = r - alpha*M@p
        r[0] = r[N] = 0
        rho_1 = np.dot(r,r)
        beta = rho_1/rho_0
        p = r + beta*p
        p[0] = p[N] = 0
        #theta = np.dot(p, mat_vec_prod(D,S,T,p))
        theta = np.dot(p,M@p)
        rho_0 = rho_1
        it = it + 1
    print ("iter", it)
    """
    """
   # BiGCStab loop 
    u = np.zeros(N+1)
    u[0]   =0
    u[N] =1
    maxiter = 2000
    it = 1
    r = f - mat_vec_prod(D,S,T,u)
    r[0]=r[N]=0
    r0_cap = r
    rho0 = alpha = omega = 1
    v = p = 0
    while (rho0>1e-12 and it<maxiter):
        rho1 = np.dot(r0_cap, r)
        beta = (rho1/rho0)*(alpha/omega)
        p = r + beta*(p - omega*v)
        p[0] = p[N] = 0 
        v = mat_vec_prod(D,S,T,u)
        v[0] = v[N] = 0
        alpha = rho1/np.dot(r0_cap, v)
        z = u + alpha*p
        s = r - alpha*v
        t = mat_vec_prod(D,S,T,s)
        t[0] = t[N] = 0
        omega = np.dot(t, s)/np.dot(t, t)
        u= h + omega*s
        u[0] =0
        u[N] =1
        r = s - omega*t 
        r[0]=r[N]=0
        rho0 = rho1
        it = it + 1
    # print ("Residual", f-mat_vec_prod(D,S,T,u))
    """
    fig,ax = plt.subplots(figsize=(8,8))
    ax.plot(x, u, 'bo-', markersize=0.1)
    ax.set_title ("Piece-wise harmonic function with a jump")
    ax.set_xlabel("x")
    ax.set_ylabel("u")
    ax.grid()
    #print ("T", T)
    vec = np.random.rand(N+1)
    #print ("check", M@vec - mat_vec_prod(D,S,T,vec))
    #print ("conditioning", np.linalg.cond(M))
    
    u, exit_code = scipy.sparse.linalg.bicgstab \
      (M, f, x0=None, tol=1e-05, maxiter=None, M=None, callback=None, atol=None)
    print ("exit code:", exit_code)
    return h, err_u_jump, err_u_der_jump #,err_u_der_jump
```

```python
def mat_vec_prod(D,S,T,u):
    b = np.zeros(len(u))
    for i in range(len(u)-1):
        sp = S[i]*(u[i+1]-u[i])
        tp = T[i]*(u[i+1]+u[i])
        b[i] = b[i]+(sp+tp)
        b[i+1] = b[i+1]-(sp+tp)
        b[i] = b[i] + D[i]*u[i]
    b[len(u)-1] = b[len(u)-1] +D[len(u)-1]*u[len(u)-1]
#    b[0]=0
#    b[len(u)-1]=0
    return b
```

```python
#check error trend
N0 =20
no_tests = 5
tab = np.zeros((3,no_tests))
for i in range (0,no_tests):
    N0 = N0*2**i
    tab [:,i] = error_vs_h(N0)
#estimate order
order = np.zeros((2,no_tests-1))
logtab = np.log(tab)
for i in range (0,no_tests-1):
    order [:,i] = (logtab[1:, i+1] -logtab[1:, i])/(logtab[0, i+1] -logtab[0, i])

fig, ax = plt.subplots(figsize=(8,8))
ax.grid(which="both")
ax.loglog (tab[0,:], tab[1,:], 'rd--', label='Order: '+str(order[0,-1]))
ax.loglog (tab[0,:], tab[2,:], 'bd--', label='Order: '+str(order[1,-1]))
ax.set_title("Convergence Order")
ax.set_xlabel("h")
ax.set_ylabel("Err")
ax.legend()


```
